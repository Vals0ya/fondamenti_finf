## get_unique(x): ritorna i valori univoci dell’attributo x

```python
def get_unique_numbers(name):

    f= open("heart1.csv", "r") #apro il file in modalita lettura
    head = f.readline() #leggo la prima riga

    head= head.strip() #rimuovo lo spazio 
    attribute_name= head.split(",")
    index =attribute_name.index(name) #è la variabile che ritorna l'indice(la posizione)
    #degli elementi trovati nella lista 

    attr =[]
    for linea in f:
        linea= linea.strip() #tolgo il cr in fondo
        valori= linea.split(",") #divido i campi
        attr.append(valori[index])


    unique = []

    for number in attr:
        if number in unique:
            continue
        else:
            unique.append(number)
    return(unique)
```

## get_x_by_y(x, y, z): ritorna una lista contenente i valori dell’attributo x quando l’attributo y vale z

```python
import csv
nome_file= "heart1.csv"
def carica_dati():
    with open(nome_file) as csvfile:
        cardio_data = csv.DictReader(csvfile) #creo un array di dizionari con csv ogni
        #riga così è un dizionario chiave valore con DictReader
        return list(cardio_data) #list mi fa aggiungere ogni 
           #elemento di cardio_data ad un array
def get_x_by_y(x,y,z):
    dati = carica_dati()
    
    valori=[] #lista vuota
    for paziente in dati:
        if paziente[y]==z:
            valori.append(paziente[x])
    return valori
```

```python
get_x_by_y("BMI","Smoking", "Yes")
```

```python
get_x_by_y("BMI","Race", "Black")
```

## media(x): ritorna la media dell’attributo x

```python
def media(name):
 
 try: #si usa per gestire le eccezioni
    lista = get_values(name) #variabile contenente una lista di tutti i  valori della colonna
    
    int_lista = [float(x) for x in lista] #converto la lista di stringhe in float

    somma = 0
    for n in int_lista:
        somma += n
    media = somma/len(int_lista)
    return  round(media) #la funzione round arrotonda la media

 except ValueError:
        print("ERRORE, DEVE ESSERE UN VALORE NUMERICO")
```

```python
media("BMI")
```

```python
media("SleepTime")
```

## get_hist(x) : ritorna una lista conente le 2 liste (l3, l4) la prima contiene i valori univoci dell’attributo x e la seconda le rispettive occorrenze

##### GET_VALUES(name): è servito a noi per avere i valori per ogni indice

```python
def get_values(name): 
    f= open("heart1.csv", "r") #apro il file
    head= f.readline() #leggo la prima riga

    head= head.strip() #rimuovo il cr
    attribute_name= head.split(",")
    index =attribute_name.index(name)

    attr =[]
    for linea in f:
        linea= linea.strip() #tolgo il cr in fondo
        valori= linea.split(",") #divido i campi
        attr.append(valori[index]) 
    f.close()#chiudo il file
    return attr
#trova l'indice dell'attributo name
```

```python
def get_hist(name):

    list1 = get_values(name)
    list2 = get_unique_numbers(name)
    list3 = []
    list4 = []

    for n in list2:
        list4.append(n)
        list3.append(list1.count(n))

    x = []
    x.append(list4)
    x.append(list3) 
    return x
```

```python
get_hist("Smoking")
```

## GRAFICI

```python
import matplotlib.pyplot as plt
```

```python
def pie (name):
    #pop serve per eliminare l'argomento messo nella parentesi e ritorna l'argomento eliminato
    x = get_hist(name).pop(1)
    labelsx = get_hist(name).pop(0)
    plt.title(name)
    plt.pie(x , labels = labelsx)
    plt.legend(x,title=name,  loc= "upper left")
    plt.show()
```

```python
pie("Smoking")
```
